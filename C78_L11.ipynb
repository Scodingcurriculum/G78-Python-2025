```python
# ==============================
# Lesson 11: Math Tutoring App
# Topics: functions (positional/keyword/default args), simple validation, while menu loop,
#         basic polymorphism (same function working with different types)
# Project Theme: Homework Helper
# Goal (Main Activity): Build a calculator with reusable functions.
# Extension: Add power and square root functions, summary stats, and practice menu.
# ==============================

print("=== MATH TUTORING APP ===")

# -------------- Core Functions --------------
def add(a, b):
    """Return a + b (works for numbers and strings -> demonstrates polymorphism)."""
    return a + b

def subtract(a, b):
    """Return a - b."""
    return a - b

def multiply(a, b=1):
    """Return a * b with a default for b."""
    return a * b

def divide(a, b):
    """Return a / b; handle division by zero simply."""
    if b == 0:
        return "Undefined (division by zero)"
    return a / b

# -------------- Input Helpers --------------
def read_float(prompt):
    while True:
        raw = input(prompt).strip()
        try:
            return float(raw)
        except:
            print("Please enter a valid number.")

# -------------- MAIN ACTIVITY --------------
# Demonstrate functions once with user inputs
x = read_float("Enter first number (x): ")
y = read_float("Enter second number (y): ")

print(f"add(x, y)              = {add(x, y)}")
print(f"subtract(x, y)         = {subtract(x, y)}")
print(f"multiply(x, y)         = {multiply(x, y)}")            # positional
print(f"multiply(a=x, b=y)     = {multiply(a=x, b=y)}")        # keyword
print(f"divide(x, y)           = {divide(x, y)}")

# Polymorphism demo: same function add() used on strings
print("Polymorphism demo:", add("Grade ", "Seven"))

# === ADDITIONAL ACTIVITY START ===
# A) Extra Functions: power and square root
def power(base, exp=2):
    """Return base ** exp (default square)."""
    return base ** exp

def sqrt(n):
    """Return square root for n >= 0."""
    if n < 0:
        return "Undefined for negative numbers (real domain)."
    return n ** 0.5

print("\nExtra functions:")
print(f"power(x, 3) = {power(x, 3)}")
print(f"sqrt(x)     = {sqrt(x)}")

# B) Simple Statistics Helpers (mean and median) using functions
def mean(a_list):
    """Compute arithmetic mean of a list of numbers."""
    if not a_list:
        return 0
    total = 0
    for val in a_list:
        total += val
    return total / len(a_list)

def median_of_three(a, b, c):
    """Return median of three numbers using simple comparisons (no sort needed)."""
    # Many correct ways exist; we keep it explicit and readable:
    if a <= b <= c or c <= b <= a:
        return b
    elif b <= a <= c or c <= a <= b:
        return a
    else:
        return c

# C) Practice Menu (while loop) â€” students can keep using the app
print("\n--- PRACTICE MENU ---")
print("Type: add, sub, mul, div, pow, sqrt, mean3, median3, or quit")
while True:
    op = input("\nOperation: ").strip().lower()
    if op == "quit":
        print("Session ended.")
        break
    elif op == "sqrt":
        n = read_float("  n: ")
        print("  Result:", sqrt(n))
    elif op == "pow":
        base = read_float("  base: ")
        exp  = read_float("  exponent: ")
        print("  Result:", power(base, exp))
    elif op == "add":
        a = read_float("  a: "); b = read_float("  b: ")
        print("  Result:", add(a, b))
    elif op == "sub":
        a = read_float("  a: "); b = read_float("  b: ")
        print("  Result:", subtract(a, b))
    elif op == "mul":
        a = read_float("  a: "); b = read_float("  b: ")
        print("  Result:", multiply(a, b))
    elif op == "div":
        a = read_float("  a: "); b = read_float("  b: ")
        print("  Result:", divide(a, b))
    elif op == "mean3":
        a = read_float("  a: "); b = read_float("  b: "); c = read_float("  c: ")
        print("  Mean:", mean([a, b, c]))
    elif op == "median3":
        a = read_float("  a: "); b = read_float("  b: "); c = read_float("  c: ")
        print("  Median:", median_of_three(a, b, c))
    else:
        print("  Unknown operation. Try again.")
```
